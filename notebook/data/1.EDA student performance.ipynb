## lets perform F.Engineering

#### add columns of total score and average

```python


df['total_score'] = df['math_score']+ df['reading_score'] + df['writing_score']
df['Average_score'] = df['total_score']/3
```

```python
df.head()
```

```python
# lets see how many students will get the full marks out of math score ,reading score and writing score
math_marks = df[df['math_score']==100]['Average_score'].count()
reading_marks = df[df['reading_score']==100]['Average_score'].count()
writing_marks = df[df['writing_score']==100]['Average_score'].count()

# lets print to them one by one
print("in math {} students get full marks".format(math_marks))
print('in reading {} students get full marks'.format(reading_marks))
print('in writing {} students get full marks'.format(writing_marks))
```

```python
math_marks = df[df['math_score'] <= 20]['Average_score'].count()
reading_marks = df[df['reading_score'] <= 20]['Average_score'].count()
writing_marks = df[df['writing_score'] <= 20]['Average_score'].count()

# lets print to them one by one
print("in math {} students get less than 20 marks out of 100".format(math_marks))
print('in reading {} students get less than 20 marks out of 100"'.format(reading_marks))
print('in writing {} students get less than 20 marks out of 100"'.format(writing_marks))
```

#### observations:- 
- 1) best performance you can see in the Reading section
- 2) and worst performance is in the math section
